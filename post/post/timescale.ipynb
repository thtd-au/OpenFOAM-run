```python
Y = ["H2O","H2","O2","OHminus","Hplus"]
D = [2.3e-9,4.5e-9,2.0e-9,5.3e-9,9.3e-9,]

L = 5.0e-5
dt = 0.005

u = 0.005 
```

```python
tau = []
a = 0
for Di in D:
    y = Y[a]
    adv = L/u
    diff = (L**2)/Di
    tau_i = min(adv,diff)
    print(f"{y}: tau = {tau_i:.4f} s, adv = {L/u:.4f} s, diff = {(L**2)/Di:.4f} s")
    tau.append(tau_i)
    a += 1
print()
print(f"smallest transport timescale tau_min = {min(tau):.4f} s")
```

```python
## Water dissociation 
pH = 7
H = 10**(-pH)
OH = 1e-14 / H

kf = 1.4e-3
kb = 1.4e11
tau_1 = 1/(kb*(H + OH))

## CO2 hydration 
kf = 3.06e-2
kb = 25.98
tau_2D = 1/(kf+kb)

c = 1e-3

## acid dissociation
kf = 8e6
kb = 4.7e10
tau_2a = 1/(kf+kb*(2*c))

## Bicarbonate carbonate
kf = 2.35
kb = 4.7e10
tau_3 = 1/(kf+kb*(2*c))

## Acetic acid acetate
kf = 2.5e5
kb = 1.4e10
tau_4 = 1/(kf+kb*(2*c))

print(f"r1: tau = {tau_1:.3g} s")
print(f"r2 d: tau = {tau_2D:.3g} s")
print(f"r2 a: tau = {tau_2a:.3g} s")
print(f"r3: tau = {tau_3:.3g} s")
print(f"r4: tau = {tau_4:.3g} s")

```
